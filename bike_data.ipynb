## Exploratory Data Analysis (EDA)

### Explore data day_df

```python
day_df.describe(include="all")
```



*   merubah kolom mth menjadi nama bulan



```python
day_df['mnth'] = day_df['dteday'].dt.month_name()
```

```python
day_df.head()
```



*   merubah kolom yr menjadi hitungan tahun



```python
day_df['yr'] = day_df['dteday'].dt.year
```

```python
day_df.head()
```



*   merubah kolom weekday menjadi nama hari



```python
day_df['weekday'] = day_df['dteday'].dt.day_name()
```

```python
day_df.head()
```



*   merubah kolom season menjadi nama-nama musim



```python
def find_season(season):
  season_str = {1:'Winter', 2:'Spring', 3:'Summer', 4:'Autumn'}
  return season_str.get(season)

season_list = []

for season in day_df['season']:
  season = find_season(season)
  season_list.append(season)

day_df['season'] = season_list
```

```python
day_df.head()
```



*   Mengetahui perbandingan antara peminjam casual dan peminjam registered



```python
total_casual = day_df['casual'].sum()
total_registered = day_df['registered'].sum()

plt.figure(figsize=(10, 7))
sns.barplot(x=["Casual", "Registered"], y=[total_casual, total_registered])
plt.xlabel("Tipe Peminjam")
plt.ylabel("Total")
plt.title("Total Peminjam Casual dan Registered")

plt.show()

print(f"Total peminjam casual: {total_casual}")
print(f"Total peminjam registerd: {total_registered}")
```



*   Mengetahui banyaknya peminjam sepeda berdasarkan musim



```python
plt.figure(figsize=(10, 7))
sns.barplot(x="season", y="cnt", data = day_df)

plt.xlabel("Season")
plt.ylabel("Total")
plt.title("Total Peminjaman Berdasarkan Season")

plt.show()
```


*  Mengetahui banyaknya peminjam sepeda berdasarkan bulan




```python
plt.figure(figsize=(10, 7))
sns.barplot(x="mnth", y="cnt", data = day_df)

plt.xlabel("Mounth")
plt.ylabel("Total")
plt.title("Total Peminjaman Berdasarkan Bulan")

plt.show()
```


*   Mengetahui banyaknya peminjam sepeda berdasarkan hari

```python
plt.figure(figsize=(10, 7))
sns.barplot(x="weekday", y="cnt", data = day_df)

plt.xlabel("weekday")
plt.ylabel("Total")
plt.title("Total Peminjaman Berdasarkan Hari")

plt.show()
```

## Visualization & Explanatory Analysis

### Pertanyaan 1: Bagaimana performa dari usaha peminjaman sepeda ini?

```python
day_df['month_year'] = day_df['dteday'].dt.strftime('%B %Y')
day_df['month_year'] = pd.Categorical(day_df['month_year'], categories=day_df['month_year'].unique(), ordered=True)

plt.figure(figsize=(10, 7))
sns.lineplot(x="month_year", y="cnt", data=day_df)
plt.xlabel("Date")
plt.ylabel("Total Peminjaman")
plt.title("Total Peminjaman Sepeda (Bulanan)")
plt.xticks(rotation=45)
plt.show()
```

### Pertanyaan 2: Apakah musim akan mempengaruhi banyaknya jumlah peminjaman sepeda? Jika Musim benar mempengaruhi banyaknya jumlah peminjaman sepeda, maka di musim apa yang memiliki jumlah peminjaman terbanyak?


```python
plt.figure(figsize=(10, 7))
sns.barplot(x="season", y="cnt", data = day_df)

plt.xlabel("Season")
plt.ylabel("Total")
plt.title("Total Peminjaman Berdasarkan Season")

plt.show()
```

### Pertanyaan 3: Apakah hari mempengaruhi banyaknya jumlah peminjaman sepeda? Jika hari benar mempengaruhi banyaknya jumlah peminjaman sepeda, umumnya pada hari apakah yang memiliki jumlah peminjam terbanyak?

```python
plt.figure(figsize=(10, 7))
sns.barplot(
        x="weekday",
        y="cnt",
        data=day_df
)
plt.title("Total Peminjaman Berdasarkan Hari", loc="center", fontsize=15)
plt.ylabel("Hari")
plt.xlabel("Total")
plt.tick_params(axis='y', labelsize=12)
plt.show()
```

## Conclusion

- Berdasarkan hasil analisis yang telah dilakukan, dapat dilihat bahwa jika dibandingkan tahun 2011, performa usaha peminjaman sepeda meningkat pada tahun 2012 dan berada dipuncaknya pada bulan Agustus 2012, namun kemudian performa peminjaman sepeda kembali mengalami penurunan.

- Melihat dari hasil data analisis, dapat diketahui bahwa musim (season) memang berpengaruh terhadap banyaknya total peminjaman sepeda. Hal tersebut dibuktikan dengan adanya perbedaan total peminjaman disetiap musimnya. Dan diantara 4 musim yang ada, peminjaman sepeda terbanyak terjadi pada summer season (musim panas).



- Melihat dari hasil data analisis, dapat dikatakan bahwa hari (weekday)cukup berpengaruh terhadap banyaknya total peminjaman sepeda. Hal tersebut dibuktikan dengan adanya selisih perbedaan total peminjaman disetiap harinya. Dan umumnya hari yang memiliki total peminjaman terbanyak adalah hari jumat (friday)
